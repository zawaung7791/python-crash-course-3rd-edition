## bicycles.py

```
# lists are created by using square brackets []
# lists are ordered, but can contain different variable types
# python doesn't have arrays
# lists are mutable

bicycles = ['trek', 'cannondale', 'redline', 'specialised']
print(bicycles)

# access elements in a list with index
print(bicycles[0])

# loop through list
for i in bicycles:
    print(i)

# slice a list, with a range
my_bike = bicycles[1:-1]
print(my_bike)

print(bicycles[0].capitalize())
print(bicycles[0].count("z"))

message = f"My first bicycle was a {bicycles[0].title()}"
print(message)
```

## motorcycles.py

```
motorcycles = ["honda", "yamaha", "suzuki"]
print(motorcycles)

motorcycles[0] = "ducati"
print(motorcycles)

# add element to list
motorcycles = ["honda", "yamaha", "suzuki"]
motorcycles.append("ducati")
print(motorcycles)

# create an empty list with just []. no values inserted
# then use append to add to list
motorcycles = []
motorcycles.append("honda")
motorcycles.append("yamaha")
motorcycles.append("suzuki")

# use insert to add values at a certain location in the list
print(motorcycles)
motorcycles.insert(0, "ducati")
print(motorcycles)

# use del to remove item from list, at a certain location
# cannot be assigned to a variable
del motorcycles[0]
print(motorcycles)

# to pop last item from the list
# or to pop from a certain index
popped_motorcycles = motorcycles.pop()
print(motorcycles)
print(popped_motorcycles)

motorcycles = ["honda", "yamaha", "suzuki"]
first_owned = motorcycles.pop(0)

print(f"The first motorcycle i owned was a {first_owned.title()}")

# remove, n.b. only removes the first oocurance
motorcycles = ["honda", "yamaha", "suzuki", "ducati"]
too_expensive = "ducati"
motorcycles.remove("ducati")
print(motorcycles)

# to completely delete the motorcycleslist
#del motorcycles
```

## 3-7 Shrinking guest list

```
# guestlist
guest_list = ["John Lennon", "Michael Stipe", "Bono", "Kurt Cobain"]

for name in guest_list:
    print(f"{name} you're invited to dinner")

print(f"\n{guest_list[-1]} can't come to dinner\n")

guest_list[-1] = "Roy Orbison"

for name in guest_list:
    print(f"{name} you're invited to dinner")

print("\nA bigger table has been found and 3 more guests invited\n")

guest_list.insert(0, "Neil Finn")
guest_list.insert(3, "Snoop Dogg")
guest_list.append("Elton John")

for name in guest_list:
    print(f"{name} you're invited to dinner")

print("\nApologies we can only invite two guests for dinner!\n")

counter = len(guest_list)

for i in range(0, counter):
    counter -= 1
    if counter == 1:
        break
    else:
        print(f"Sorry {guest_list[counter]} you can't come to dinner")
        del guest_list[counter]

for name in guest_list:
    print(f"Don't worry {name} you're still invited for dinner")    

guest_list.clear()
print(guest_list)
```

## cars.py

```
# sorting list permanently
cars = ["bmw", "audi", "porsche", "bentley", "ford", "vauxhall"]
print(cars)

cars.sort()
print(cars)

# sorting in reverse permanently
cars.sort(reverse=True)
print(cars)

# temoprarily sorting a list
cars = ["bmw", "audi", "porsche", "bentley", "ford", "vauxhall"]
print("Here's the original list\n", cars)
print("Here's the sorted list\n", sorted(cars))
print("Here's the original list\n", cars)
print("Here's the reversed list\n", sorted(cars, reverse=True))
print("Here's the original list\n", cars)

# reverse the list not alphabetical, permantely
cars.reverse()
print("Cars in reverse", cars)

```

## 3-10 Every function

```
my_list = ["lfc", "tomatos", "kfc", "rem", "hte"]

print(f"The original list\n{my_list}\n")

# append
my_list.append("htaik")
print(f"Modified list\n{my_list}\n")

# copy my_list to new_list
new_list = my_list.copy()
print(f"Copied list\n{new_list}\n")

# count 
my_list.count("tomatos")

# extend list by characters, permanetly
my_list.extend("pudding")
print(my_list)

# index, find index of string
my_list.index("hte")

# insert at certain index
my_list.insert(0, "guitar")
print(my_list)

# remove last item permenately
my_list.pop()
print(my_list)

# remove by item value
my_list.remove("guitar")
print(my_list)

# reverse non alphabetical
my_list.reverse()
print(my_list)

# sort, alphabetical or not permenant
my_list.sort()
print(my_list)

my_list.sort(reverse=True)
print(my_list)

# sort temoprarily
print(sorted(my_list))

# del element
del my_list[-1]
print(my_list)

# clear list
my_list.clear()
print(my_list)

# delete list
del my_list
```
