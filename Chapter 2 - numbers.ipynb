# Chapter 2 - Numbers

## Integers

```
print(3*2)
print(3+2)
print(3-2)

# division of integers always results in floats
print(3/2)

# to get integer division use //
print(3//2)

# use ** for exponent
print(3 ** 2)
```

## Floats

```
print(0.2 + 0.1)

# if you mix integers and flotas, always result will be a float

print(1 + 2.0)
print(2 * 3.0)
print(3 / 3)

# use underscores in numbers, to make large numbers readable
universe_age = 14_000_000_000

print(universe_age)

# multiple assignment
x, y, z = 0, 1, 2

print(x, y, z)

# no built-in constants, use capital letters to highlight
MAX_CONNECTIONS = 500

```
