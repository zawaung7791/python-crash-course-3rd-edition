## full-name.py

```
# use f strings to format text in strings, be careful of whitespace in string  
first_name = "ada"
last_name = "lovelace"
full_name = f"{first_name} {last_name}"

# with f strings
print(f"{first_name} {last_name}")
print(full_name)
print(f"Hello {full_name.title()}!")

# without f strings
print(first_name.title(), last_name.title(), sep=" ")
print(full_name.title())

# assign f strings to variable
message = f"Hello {full_name.title()}!"
print(message)
```

## Escape characters

```
# using \n to make new line
print("Languages:\nPython\nJavascript\nC")

# using \t to make tabspaces
print("Languages:\tPython\tJavascript\tC")

# combination of \n and \t
print("Languages:\n\tPython\n\tJavascript\n\tC")
```

## Stripping whitespace

```
# use rstrip (right strip), lstrip (left strip) and strip (both sides) to remove whitespace from text
whitespace_text_1 = "python "
whitespace_text_2 = " python"
whitespace_text_3 = " python "

print(whitespace_text_1.rstrip())
print(whitespace_text_2.lstrip())
print(whitespace_text_3.strip())
```

## apostrophe.py

```
# using double quotation marks to delimit text, can use apostrophe
message_3 = "One of Python's strengths is it's diverse comunity"
print(message_3)

# using apostrophe to delimit text, cannot use ' but " is allowed
# this will fail with a SyntaxError
message_4 = 'One of Python's strengths is it's diverse comunity'
print(message_4)
```

## 2-4 Name-cases

```
# using f strings
print(f"Hello {message_4.title()}")
print(f"Hello {message_4.upper()}")
print(f"Hello {message_4.lower()}")

# non f string 
print("Hello", message_4.title())
print("Hello", message_4.upper())
print("Hello", message_4.lower())
```

## 2-8 File Extensions and removing pre and suffixes

```
# remove prefix

test_url = "https://www.google.com"

print(test_url.removeprefix("https://www."))

# remove suffix

test_filename = "python_notes.txt"

print(test_filename.removesuffix(".txt"))

# you have to assign to new variables for changes to take place, as strings are immutable

new_test_variable = print(test_url.removeprefix("https://www."))
new_filename = print(test_filename.removesuffix(".txt"))
```
